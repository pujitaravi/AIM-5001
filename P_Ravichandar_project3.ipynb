### Removing the Header

To facilitate the parsing through the text file, the header was be removed. The approach uses indexing to remove the header. Variable called `noheader` contains the updated string.  

```python
#getting rid of new lines in the file
with open(r'/Users/Puji/Documents/AIM_5001/tournamentinfo.txt')as file:
    data= file.read().replace('\n', ' ')

#removing headers from original data
noheader = data[361:]
noheader
```

## Step 2

 * To further facilitate the parsing through the text file, special characters "-", "|", and "/" were removed. 
 * Double empty spaces were replaced by single spaces.

**Finding the values**<br>

To find the elements in the string, the regex functions re.match and re.split were utilized.

**Variables Descriptions:**

* "player_id" = Contains a list of each of the player's ids 
* "player_name" = Contains each of the player's names 
* "total_points" = Contains the total points per player, ranging from **6.0** as the highest and **1.0** as the lowest points. The scores follow increments of 0.5.

```python
#remove the dashes, bars, and slashes from the dataset
#the resulting dataset is titled data5
data2 = noheader.translate({ord(x): None for x in'-'})
data3 = data2.translate({ord(x): None for x in '|'})
data4 = data3.translate({ord(x): None for x in '/'})
data5 = re.sub('\s+',' ', data4)

#find player id
player_id = [a for a in re.split('[^0-9-()\n]', data5) if re.match('(\w+.).(\w+.)',a)]

#find player name
player_name = [a for a in re.split(r'[^A-Z+]+[^A-Z+]', data5) if re.match('(\w+.).(\w+.)',a)]

#find total points for each game
total_points = [a for a in re.split(r'[^0-9\.?0-9]', data5) if re.match('(\d\.\d)',a)]
```

## Step 3 

**Finding the state**

To parse the text and locate the states, a list comprehension was used. Elements were matched and retrieved using the regular expression functions re.split() and re.match(). A for loop iteration appended results. The variable called `state` contains each player's state of residency. 

```python
#split data2 at the bars and include instances of a space followed by two capital letters
pre_state = [a for a in re.split('\|', data2) if re.match('\s+[A-Z]{2}', a)]

#make a list of everyother element in pre_state to pull out the state values
state = []
for i in range(1, len(pre_state), 2):
    state.append(pre_state[i])

state
```

## Step 4: Pre-Rating score

Finding the pre-rating score for each player is given by isolating the desired information from the data value titled `data5`. This effort took various iterations of using regular expressions. This was mainly due to the complexity and nonuniform documentation of the pre-rating scores. Some of the scores were just numbers, whereas others included non-digit characters irrelevant to the value needed for our purpose. The specific code used for isolating the pre-rating scores is detailed below. And the final scores are listed in the `pre_rating` list object.


```python
#split the compiled data by the > character
split = re.split('\>', data5)

#remove the characters P followed by any numbers and add to the list split_sub
split_sub = []

for s in split:
    i = re.sub('[P]+\d+', '', s)
    split_sub.append(i)

#take the last five digts of split_sum
#replace any non-digit characters with nothing
#add these to a list called pre_rating_trim_x
pre_rating_trim_x = []
for s in split_sub:
    x = s[-5:]
    y = re.sub('\D', '', x)
    pre_rating_trim_x.append(y)

#cut out the last item of pre_rating_trim_x
pre_rating_trim = pre_rating_trim_x[:-1]

#pre_rating values are the integers values of the pre_rating_trim string
pre_rating = []

for p in pre_rating_trim:
    pre_rating.append(int(p))

#print pre-rating values
pre_rating
```
