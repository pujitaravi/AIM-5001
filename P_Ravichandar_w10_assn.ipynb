## Importing and Data Set Up
Approach: Create an excel sheet with the necessary data values and export it as a .csv file. Then, push the file onto the github repository. From there, the file can be read into the Jupyter notebook using the read_csv() function.

```python
#import pandas package
import pandas as pd
```

```python
#create dataframe by reading the csv uploaded to github
df = pd.read_csv("https://raw.githubusercontent.com/pujitaravi/AIM-5001/master/wk10_assn.csv")
df
```

## TIdy the Data
Approach: To tidy this dataset, the wide formate of the dataset needs to be converted to the long orientation. Looking at the resulting dataframe, the airline names need to be present for each row value, and the rows with null values can be dropped as they add no value and are not part of an observation. The final step in tidying this dataframe lies in renaming the columns to more intuative and relevant titles. The code detailing these procedures is below.

```python
#convert from wide to long format
long_df = pd.melt(df, id_vars=['Unnamed: 0', 'Unnamed: 1'])
long_df
```

```python
#fill in the relevant airlines names for the null values 
long_df['Unnamed: 0'].fillna(method='ffill', inplace = True)
long_df
```

```python
#drop rwos with null values
long_df = long_df.dropna()
long_df
```

```python
#finish the process to tidy the long format by renamming the columns
long_df_tidy = long_df.rename(columns={'Unnamed: 0': 'Airline', 'Unnamed: 1':'Arrival Delays', 
                        'variable': 'City', 'value': 'Count'})

long_df_tidy
```

## Analyze Results

### For each city, which airline had the best on time performance?
Approach: Organizing the data in the tidied dataframe is matter of grouping the values in an order that makes sense. For this question, the data should be grouped by city, arrival delays, and then the airline. From the resulting values the cities with airlines that had the best on time performance are as follows:
* Los Angeles: AM WEST
* Pheonix: AM WEST
* San Diego: AM WEST
* San Francisco: ALASKA
* Seattle: ALASKA

The code supporting this conclusion is below.

```python
#group the tidy dataframe by ciry, arrival delay, and lastly airline
city_on_time = long_df_tidy.groupby(['City', 'Arrival Delays', 'Airline'])

#sum the grouped values
city_on_time.sum()
```

### Which airline had the best overall on time performance?
Approach: The grouping to best answer this question is by airline and then by arrival delays. The results for which airline had the best overall on time performance is AM WEST. The code supporting this conclusion is below.

```python
#group the tidy datafram by airline and then arrival delay
on_time = long_df_tidy.groupby(['Airline', 'Arrival Delays'])

#sum the values in the grouped database
on_time.sum()
```
